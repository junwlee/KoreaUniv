```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# 플롯을 그리는 함수
def plot_data(correlation, ax, title):
    # 데이터 생성: n_features는 변수의 수, noise는 데이터의 노이즈 정도
    X, y = make_regression(n_samples=100, n_features=1, noise=10*(1-correlation), bias=0.5, random_state=42)
    ax.scatter(X[:, 0], y, alpha=0.6, edgecolors='w')
    ax.set_title(title)
    ax.set_xlabel('Feature X')
    ax.set_ylabel('Target Y')

fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# 상관계수 0.9, 0.5, 0.1에 대한 데이터 생성 및 플롯
plot_data(0.9, axs[0], 'High Correlation (0.9)')
plot_data(0.5, axs[1], 'Moderate Correlation (0.5)')
plot_data(0.1, axs[2], 'Low Correlation (0.1)')

plt.show()
```

- 상관계수가 0.9 이상일 때, 점들은 매우 강한 선형적 관계를 보이며, 거의 하나의 직선 주변에 모여 있다.
