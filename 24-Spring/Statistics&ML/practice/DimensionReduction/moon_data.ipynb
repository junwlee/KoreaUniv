```python
from sklearn.datasets import make_moons
import plotly.graph_objects as go

X, y = make_moons(n_samples=100, random_state=123)

fig = go.Figure()
fig.add_trace(go.Scatter(x=X[y==0,0], y=X[y==0,1], mode='markers', name='group1'))
fig.add_trace(go.Scatter(x=X[y==1,0], y=X[y==1,1], mode='markers', name='group1'))
fig.update_xaxes(title_text='x1')
fig.update_yaxes(title_text='x2')
fig.update_layout(width=600, height=600)
fig.show()
```

### X[y==0,0]과 X[y==0,1]의 의미
  - y==0: 이 조건은 y 배열에서 값이 0인 모든 인덱스를 찾습니다. 여기서 값 0은 첫 번째 그룹을 의미합니다.
  - X[y==0,0]: 이 표현은 X 배열에서 y==0 조건을 만족하는 모든 행의 첫 번째 열을 선택합니다. 첫 번째 열은 x축 좌표를 나타냅니다. 즉, 이것은 첫 번째 그룹에 속하는 모든 데이터 포인트의 x축 좌표입니다.
  - X[y==0,1]: 이 표현은 X 배열에서 y==0 조건을 만족하는 모든 행의 두 번째 열을 선택합니다. 두 번째 열은 y축 좌표를 나타냅니다. 즉, 이것은 첫 번째 그룹에 속하는 모든 데이터 포인트의 y축 좌표입니다.

### X[y==1,0]과 X[y==1,1]의 의미
  - y==1: 이 조건은 y 배열에서 값이 1인 모든 인덱스를 찾습니다. 여기서 값 1은 두 번째 그룹을 의미합니다.
  - X[y==1,0]: 이 표현은 X 배열에서 y==1 조건을 만족하는 모든 행의 첫 번째 열을 선택합니다. 첫 번째 열은 x축 좌표를 나타냅니다. 즉, 이것은 두 번째 그룹에 속하는 모든 데이터 포인트의 x축 좌표입니다.
  - X[y==1,1]: 이 표현은 X 배열에서 y==1 조건을 만족하는 모든 행의 두 번째 열을 선택합니다. 두 번째 열은 y축 좌표를 나타냅니다. 즉, 이것은 두 번째 그룹에 속하는 모든 데이터 포인트의 y축 좌표입니다.

```python
from sklearn.decomposition import KernelPCA

kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)

fig = go.Figure()
fig.add_trace(go.Scatter(x=X_kpca[y==0,0], y=X_kpca[y==0,1], mode='markers', name='group1'))
fig.add_trace(go.Scatter(x=X_kpca[y==1,0], y=X_kpca[y==1,1], mode='markers', name='group1'))
fig.update_xaxes(title_text='PCA1')
fig.update_yaxes(title_text='PCA2')
fig.update_layout(width=600, height=600)
fig.show()
```
