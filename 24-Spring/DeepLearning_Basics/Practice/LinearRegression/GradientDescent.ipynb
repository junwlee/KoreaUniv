```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0.0, 1.0, 2.0])
y = np.array([3.0, 3.5, 5.5])

W = 0
b = 0

lr = 0.01
epochs = 1000

n = float(len(X))

for _ in range(epochs):
    y_pred = W*X + b
    dW = (2 / n) * np.dot(X, (y_pred - y))
    db = (2 / n) * sum(y_pred - y)
    W -= lr * dW
    b -= lr * db
    
print(W, b)
```

```python
y_pred = W * X + b

plt.scatter(X, y, color='red')
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='blue')
plt.show()
```
