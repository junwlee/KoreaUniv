```python
import matplotlib.pylab as plt
import numpy as np
from sklearn import linear_model as lm
from sklearn.model_selection import train_test_split
from sklearn import datasets
# R^2 점수를 계산하기 위해 필요한 함수를 불러옵니다.
from sklearn.metrics import r2_score
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

bmi = diabetes_X[:,2, np.newaxis]
# diabetes_X 배열에서 모든 행(:)에 대해 세 번째 독립 변수(BMI)를 선택하고(2), 이를 2차원 배열의 형태로 만들기 위해 새로운 축을 추가

X_train, X_test, y_train, y_test = train_test_split(bmi, diabetes_y, test_size=0.2, random_state=9)

reg = lm.LinearRegression()
reg.fit(X_train, y_train)

y_train_pred = reg.predict(X_train)
y_test_pred = reg.predict(X_test)

# R^2 점수를 출력합니다.
print(r2_score(y_train, y_train_pred))
print(r2_score(y_test, y_test_pred))

plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_test_pred, color='blue', linewidth=3)
plt.show()
```

### 넘파이 배열에서 콜론은 '모든 요소'를 선택하겠다는 의미
- array[2, :] 또는 array[2]: 2차원 배열에서 세 번째 행의 모든 열을 선택

### np.newaxis는 넘파이 배열의 차원을 늘리는 데 사용

### random_state의 역할:
- random_state 값은 데이터가 분할되는 방식에 영향을 주며, 이는 훈련 데이터와 테스트 데이터의 특정 샘플링에 영향을 준다.
- random_state=9는 이 데이터셋에 대해 훈련 세트와 테스트 세트가 잘 균형을 이루도록 하여, 모델이 더 일반화된 패턴을 학습하고, 결과적으로 테스트 세트에 대한 예측 성능이 향상되는 샘플링을 만들었을 수 있다.

### R^2 점수
- 데이터에 모델이 얼마나 잘 맞는지를 나타내는 지표
- 1에 가까울수록 모델이 데이터를 잘 설명한다는 의미
