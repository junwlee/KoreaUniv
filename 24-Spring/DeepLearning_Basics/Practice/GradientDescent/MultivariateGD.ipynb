```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(x, y):
    return x**2 + y**2

def differentiation_x(x, y, delta):
    x_delta = func(x + delta, y)
    func_x = func(x, y)
    return (x_delta - func_x) / delta

def differentiation_y(x, y, delta):
    y_delta = func(x, y + delta)
    func_y = func(x, y)
    return (y_delta - func_y) / delta

delta = 1e-10
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(x, y)
Z = func(X, Y) # func 함수에 X, Y를 사용
U = -differentiation_x(X, Y, delta) # minus sign for optimize loss function
V = -differentiation_y(X, Y, delta)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

plt.figure()
Q = plt.quiver(X, Y, U, V, units='width') # (U, V) 벡터를 그래프에 표시

plt.show()
```

```python
nx, ny = (2, 3)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
xv, yv = np.meshgrid(x, y)

print(xv)
print(yv)

```

## np.meshgrid(x, y):
- 두 매트릭스 xv와 yv가 생성
    - xv는 x 벡터의 각 값이 y의 모든 값에 대해 반복되어 행 방향으로 배열됩니다.
    - yv는 y 벡터의 각 값이 x의 모든 값에 대해 반복되어 열 방향으로 배열됩니다.
