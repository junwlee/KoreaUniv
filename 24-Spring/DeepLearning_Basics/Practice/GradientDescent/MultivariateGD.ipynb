```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(x, y):
    return x**2 + y**2

def differentiation_x(x, y, delta):
    x_delta = func(x + delta, y)
    func_x = func(x, y)
    return (x_delta - func_x) / delta

def differentiation_y(x, y, delta):
    y_delta = func(x, y + delta)
    func_y = func(x, y)
    return (y_delta - func_y) / delta

delta = 1e-10
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(x, y)
Z = func(X, Y) # func  X, Y 
U = -differentiation_x(X, Y, delta) # minus sign for optimize loss function
V = -differentiation_y(X, Y, delta)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

plt.figure()
Q = plt.quiver(X, Y, U, V, units='width') # (U, V)   

plt.show()
```
