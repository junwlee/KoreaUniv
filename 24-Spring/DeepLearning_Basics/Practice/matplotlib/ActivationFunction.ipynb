# Sigmoid

```python
import numpy as np
import matplotlib.pylab as plt
```

# sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))    

X = np.linspace(-10, 10, 4000)
Y = sigmoid(X)
plt.plot(X, Y)
plt.ylim(-0.1, 1.1)
plt.grid(True)  # 격자 추가
plt.title('sigmoid_2016260303_LeeJunWon', fontsize=20, color='blue', loc='center')
plt.show()
```

# tanh

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

X = np.linspace(-10, 10, 4000)
Y = tanh(X)

plt.plot(X, Y)
plt.ylim(-1.1, 1.1)
plt.grid(True)  # 격자 추가
plt.scatter(0, 0, color='red')  # (0,0)에 빨간 점 추가
plt.title('tanh_2016260303_LeeJunWon', fontsize=20, color='blue', loc='center')
plt.show()
```

# ReLU

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 4000)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 10.5)
plt.grid(True)  # 격자 추가
plt.title('ReLU_2016260303_LeeJunWon', fontsize=20, color='blue', loc='center')
plt.show()
```

# Leaky ReLU

```python
def leaky_relu(x):
    return np.maximum(0.1 * x, x)

x = np.linspace(-10, 10, 4000)
y = leaky_relu(x)
plt.plot(x, y)
plt.ylim(-5.0, 10.5)
plt.grid(True)  # 격자 추가
plt.title('Leaky ReLU_2016260303_LeeJunWon', fontsize=15, color='blue', loc='center')
plt.show()
```

# ELU

```python
def elu(x, alpha):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

x = np.linspace(-10, 10, 4000)
alpha = 1.0
y = elu(x, alpha)
plt.plot(x, y)
plt.ylim(-5.0, 10.5)
plt.grid(True)  # 격자 추가
plt.title('ELU_2016260303_LeeJunWon', fontsize=15, color='blue', loc='center')
plt.show()
```
