```python
import pandas as pd

amzn_data = pd.read_csv("data/AMZN.csv", header=0)
amzn_data['Date'] = pd.to_datetime(amzn_data['Date'])  # Date 열을 datetime 형식으로 변환
amzn_data.set_index('Date', inplace=True)              # Date 열을 인덱스로 설정
```

```python
import matplotlib.pyplot as plt

# Calculate moving averages (5-day, 10-day, 20-day)
amzn_data['5-day MA'] = amzn_data['Close'].rolling(window=5).mean()
amzn_data['10-day MA'] = amzn_data['Close'].rolling(window=10).mean()
amzn_data['20-day MA'] = amzn_data['Close'].rolling(window=20).mean()

# Plot the Close price along with the moving averages
plt.figure(figsize=(14, 7))
plt.plot(amzn_data['Close'], label='Close Price', color='blue', linewidth=1)
plt.plot(amzn_data['5-day MA'], label='5-Day MA', color='orange', linestyle='--')
plt.plot(amzn_data['10-day MA'], label='10-Day MA', color='green', linestyle='--')
plt.plot(amzn_data['20-day MA'], label='20-Day MA', color='red', linestyle='--')

plt.title('Amazon Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
